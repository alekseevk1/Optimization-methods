**Метод Брента.**

```python
import numpy as np
import matplotlib.pyplot as plt

from brents_method import one_dim
from result import Result
```

```python
N = 15
epsilons = 10.0 ** (-1 * np.arange(1, N))
params = []
funcs = []

params.append([-4.5, -1])
funcs.append(lambda x: np.sin(x + 1) / x / x / x / x)

params.append([1, 9])
funcs.append(lambda x: (x - 4) ** 2 + 5)

params.append([0.1, 1])
funcs.append(lambda x: np.exp(x ** 2 - 4 * np.sin(x)) + x ** ((np.log(x) + 2) / x))

params.append([0.99, 5])
funcs.append(lambda x: (np.power(x, 0.25)) * (np.e ** (1 / x)))
```

```python
for param, func in zip(params, funcs):
    times = []
    iterations = []
    for eps in epsilons:
        res = one_dim(param[0], param[1], func, eps)
        times.append(res.time)
        iterations.append(res.iterations)
        
    plt.figure(figsize=(15, 8))
    plt.plot(np.arange(1, N), times, label="times, sec", color="blue")
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(15, 8))
    plt.plot(np.arange(1, N), iterations, label="iterarions, n", color="orange")
    plt.legend()
    plt.show()
    
```

```python
for param, func in zip(params, funcs):
    res = one_dim(param[0], param[1], func)
    print(res.x)
    plt.figure(figsize=(15, 8))
    grid = np.linspace(param[0], param[1], 1000)
    plt.plot(grid, func(grid), color="blue", label="func_name")
    plt.plot(res.points, res.values, alpha=0.8, color="red", label="points")
    plt.scatter(res.points, res.values, color="black", alpha=0.7)
    plt.legend()
    plt.show()
    
```
